```python
#Comparison Plot between REINFORCE AND REINFORCE WITH BASELINE+SAMPLING+CAUSALITY
#Refer Cumulative_Results_Final.csv for reward details

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('Cumulative_Results_Final.csv')

plt.figure(figsize = (8,6))
plt.plot(df['Iteration'],df['OLD_REINFORCE_reward'],color = 'r',label = 'REINFORCE')
plt.plot(df['Iteration'],df['REINFORCE_reward'],color = 'b',label = 'CA+SA+BA REINFORCE')
plt.xlabel('No of batches/epochs')
plt.ylabel('Average Batch Reward')
plt.title('REINFORCE vs REINFORCE with baseline,sampling & causality (Learning rate = 0.01,Epsilon=0)')
plt.legend(loc='best')
```

```python
#Ablation_Study
df1 = pd.read_csv('AblationStudy_Final.csv')

plt.figure(figsize = (8,6))
plt.plot(df1['Iteration'],df1['BS_IS_REINFORCE_reward'],color = 'r',label = 'Baseline+Important Sampling')
plt.plot(df1['Iteration'],df1['BS_CA_REINFORCE_reward'],color = 'b',label = 'Baseline+Causality')
plt.plot(df1['Iteration'],df1['CA_IS_REINFORCE_reward'],color = 'm',label = 'Important Sampling+Causality')
plt.plot(df['Iteration'],df['REINFORCE_reward'],color = 'g',label = 'All three')
plt.xlim((0,100))
plt.xlabel('No of batches/epochs')
plt.ylabel('Average Batch Reward')
plt.title('Ablation Study (Learning rate = 0.01,Epsilon=0)')
plt.legend(loc='best')
```
