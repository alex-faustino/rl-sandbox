## Reinforce Algorithm

Derivation for Reinforce Algorithm for Tabular case

let $\theta_{sa}$ be the weights parameterizing each action $a$ in each state $s$, we can then define $\Theta$ as follows

$ \Theta  = \begin{bmatrix}
    \theta_{11} & \theta_{12} & \dots  & \theta_{1m} \\
    \theta_{21} & \theta_{22} & \dots  & \theta_{2m} \\
    \vdots & \vdots & \ddots & \vdots \\
    \theta_{n1} & \theta_{n2}  & \dots  & \theta_{nm}
\end{bmatrix}$

where the system has $n$ discrete states and $m$ discrete actions.

Probability mass functions for each action given states and parameter $\theta$, can be obtained using the softmax fuction

\begin{align}
P(a|s;\theta) = \frac{e^{\theta_{as}}}{\sum_{a' = 1}^m e^{\theta_{a's}}}\\
ln(P(a|s;\theta)) = \theta_{as} - ln{\sum_{a' = 1}^m e^{\theta_{a's}}}\\
\nabla ln(p) = \delta_{ss} \dot [ \delta_{aa} - P(a|s;\theta) ]
\end{align}
