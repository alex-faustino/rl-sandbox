```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# File Name for saving the Results to file
file_name = 'hw4_PGgridWorldmlp_easyon'

with open(file_name,'rb') as file:
    PG_avgRewardmlp_easyon = pickle.load(file)
    
file_name = 'hw5_bias_PGgridWorldmlp_easyon'

with open(file_name,'rb') as file:
    baseline_avgRewardmlp_easyon = pickle.load(file)
    
file_name = 'hw5_IS_PGgridWorldmlp_easyon'
with open(file_name,'rb') as file:
    IS_avgRewardmlp_easyon = pickle.load(file)
    
file_name = 'hw5_PGgridWorld_Causal_easyon'
with open(file_name,'rb') as file:
    Casual_avgRewardmlp_easyon = pickle.load(file)
    

default, = plt.plot(PG_avgRewardmlp_easyon,label = 'PG-Causal')
baseline, = plt.plot(baseline_avgRewardmlp_easyon, label='PG-Baseline')
importance_sampling, = plt.plot(IS_avgRewardmlp_easyon, label = 'PG-IS')
Causal, = plt.plot(Casual_avgRewardmlp_easyon, label = 'PG-non-Causal')
plt.grid(axis='both')
plt.xlabel('Episodes')
plt.ylabel('Avg Reward')
plt.title('Deterministic-Grid world')
plt.legend(handles=[default,baseline,importance_sampling,Causal])
plt.show()
    
```
