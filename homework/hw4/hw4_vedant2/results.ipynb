- def QLearn(env,initial_epsilon = 1, final_epsilon = 0.01,total_episodes = 1000, annealing_period = None,max_steps = 25,lr_rate = 0.9,gamma = 0.9, decay_rate = None):

- def Reinforce(env,initial_epsilon = 1, final_epsilon = 0.01,total_episodes = 1000, annealing_period = None,max_steps = 25,lr_rate = 0.9,gamma = 0.9, decay_rate = None):

if decay_rate is not none, exponential annealing model is used with the following formula:

epsilon = initial_epsilon + (final_epsilon - initial_epsilon) * np.exp(-decay_rate * episode) 

## SARSA

SARSA agent for Gridworld

### GridWorld (easy)

```python
# -*- coding: utf-8 -*-
"""
Created on Wed Sep  5 20:30:50 2018

@author: Vedant
"""
import Agents.discreteAgent as DA
import gym
import gridWorld
import numpy as np


env = gym.make('GridWorld-v0')
#env.render('human')
Q, rewards = DA.SARSA(env,initial_epsilon = 1, final_epsilon = 0.01,total_episodes = 1000)
V = np.amax(Q, axis=1)
V = V.reshape(5,5)
pi = np.argmax(Q, axis=1)
pi = pi.reshape(5,5)
print("learnt Q")
print(V)
print("learnt policy")
print(pi)
```

## Reinforce Algorithm

Derivation for Reinforce Algorithm for Tabular case

let $\theta_{sa}$ be the weights parameterizing each action $a$ in each state $s$, we can then define $\Theta$ as follows

$ \Theta  = \begin{bmatrix}
    \theta_{11} & \theta_{12} & \dots  & \theta_{1m} \\
    \theta_{21} & \theta_{22} & \dots  & \theta_{2m} \\
    \vdots & \vdots & \ddots & \vdots \\
    \theta_{n1} & \theta_{n2}  & \dots  & \theta_{nm}
\end{bmatrix}$

where the system has $n$ discrete states and $m$ discrete actions.

Probability mass functions for each action given states and parameter $\theta$, can be obtained using the softmax fuction

\begin{align}
P(a|s;\theta) = \frac{e^{\theta_{as}}}{\sum_{a' = 1}^m e^{\theta_{a's}}}\\
ln(P(a|s;\theta)) = \theta_{as} - ln{\sum_{a' = 1}^m e^{\theta_{a's}}}\\
\nabla ln(p) = \delta_{ss} \dot [ \delta_{aa} - P(a|s;\theta) ]
\end{align}
