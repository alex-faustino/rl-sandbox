```python
import gym
import gym.spaces
import matplotlib.pyplot as plt
from PPO import PPO
%matplotlib inline
```

Create gym environment

```python
env = gym.make('Pendulum-v0')
env.seed()
```

Hyperparameters

```python
gamma = 0.9
clip = 0.2
max_grad_norm = 0.5
ppo_epoch = 10
buffer_size = 1000
batch_size = 32
```

Training parameters

```python
episodes_num = 1000
episode_length = 150
```

Create agent

```python
ppo = PPO(env, gamma, clip, max_grad_norm, ppo_epoch, buffer_size, batch_size)
```

Train the agent and plot the learning curve

```python
all_cum_rewards = ppo.train(episodes_num, episode_length, False)

plt.plot(range(episodes_num), all_cum_rewards, 'r')
plt.ylabel('episode cumulative reward')
plt.xlabel('episode number')
plt.show()
```
