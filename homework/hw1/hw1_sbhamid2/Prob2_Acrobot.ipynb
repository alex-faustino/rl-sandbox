```python
import gym
import matplotlib.pyplot as plt
import random
import numpy as np
from acrobot.acrobot import AcroEnvNew

env = gym.make('MyAcrobot-v2')
store_state = {}
store_action = {}
store_reward = {}

env.seed(100)
for i_episode in range(20):
    store_state[str(i_episode)] = []
    store_action[str(i_episode)] = []
    store_reward[str(i_episode)] = []

    init_state = env.reset()
    print('Reset state: ', init_state)
    for t in range(100):
        env.render()
        action = random.randint(0,2)-1 #env.action_space.sample()-1
        store_action[str(i_episode)].append(action)

        observation, reward, done, info = env.step(action)
        store_state[str(i_episode)].append(observation)
        store_reward[str(i_episode)].append(reward)
        print(" ")
        print("Action: ", action, "Reward: ", reward)   

        if done:
            print("Episode finished after {} timesteps".format(t+1))
            break
env.close()
```

```python
idx = random.randint(0,1)
print('idx: ', idx)
plt.plot(store_reward[str(idx)], 'b')
plt.plot(store_reward[str(idx)], 'b.')
plt.xlabel('Iterations')
plt.title('Reward obtained at each instant')
plt.ylabel('Reward')
plt.show()
```
