```python
import gym
import matplotlib.pyplot as plt
import random
import numpy as np
from gridworld.gridworld import GridWorldEnvNew

env = gym.make('MyGridworld-v2')
store_state = {}
store_action = {}
store_reward = {}

env.seed(100)
for i_episode in range(20):
    store_state[str(i_episode)] = []
    store_action[str(i_episode)] = []
    store_reward[str(i_episode)] = []
    
    init_state = env.reset()
    print('Reset state: ', init_state)
    for t in range(1000):
        act_x = random.randint(0,2)-1 #gym.spaces.Discrete(2).sample()
        if act_x ==0:
            act_y = 2*random.randint(0,1)-1
        else: 
            act_y = 0
        action = [act_x, act_y]
        store_action[str(i_episode)].append(action)
        observation, reward, done, info = env.step(action)
        print(" ")
        print("Action: ", action)     
        print("Reward: ", reward)    
        store_state[str(i_episode)].append(observation)
        store_reward[str(i_episode)].append(reward)
        env.render()
```

```python
idx = random.randint(0,1)
print('idx: ', idx)
plt.plot(store_reward[str(idx)], 'b')
plt.plot(store_reward[str(idx)], 'b.')
plt.xlabel('Iterations')
plt.ylabel('Reward')
plt.title('Reward obtained at each instant')
```

```python
plt.subplot(211)
plt.plot(np.transpose(store_state[str(idx)])[0], 'b')
plt.plot(np.transpose(store_state[str(idx)])[0], 'b.')
plt.ylabel('X-State')
plt.title('State/Observation')
plt.subplot(212)
plt.plot(np.transpose(store_state[str(idx)])[1], 'b')
plt.plot(np.transpose(store_state[str(idx)])[1], 'b.')
plt.ylabel('Y-State')
plt.xlabel('Iterations')
```

```python
plt.subplot(211)
plt.plot(np.transpose(store_action[str(idx)])[0], 'b')
plt.plot(np.transpose(store_action[str(idx)])[0], 'b.')
plt.ylabel('X-Action')
plt.title('Random Action')
plt.subplot(212)
plt.plot(np.transpose(store_action[str(idx)])[1], 'b')
plt.plot(np.transpose(store_action[str(idx)])[1], 'b.')
plt.ylabel('Y-Action')
plt.xlabel('Iterations')
```
