```python
import gym
import matplotlib.pyplot as plt
import random
import numpy as np

env = gym.make('MyLocalizeRobot-v1')
store_state = {}
store_action = {}
store_reward = {}
store_meas = {}

env.seed(100)
for i_episode in range(2):
    store_state[str(i_episode)] = []
    store_action[str(i_episode)] = []
    store_reward[str(i_episode)] = []
    store_meas[str(i_episode)] = []
    
    init_state = env.reset()
    print('Reset state: ', init_state)
    for t in range(1000):
        act_x = random.randint(0,2)-1 #gym.spaces.Discrete(2).sample()
        if act_x ==0:
            act_y = 2*random.randint(0,1)-1
        else: 
            act_y = 0
        action = [act_x, act_y]
        store_action[str(i_episode)].append(action)
        meas, reward, done, error = env.step(action)
        print(" ")
        print("Action: ", action)     
        print("Reward: ", reward)   
        print("State Error: ", error)   
        
        store_state[str(i_episode)].append(error)
        store_meas[str(i_episode)].append(meas)
        store_reward[str(i_episode)].append(reward)
        env.render()

env.close()
```

```python
idx = random.randint(0,1)
print('idx: ', idx)
plt.plot(store_reward[str(idx)], 'b')
plt.plot(store_reward[str(idx)], 'b.')
plt.xlabel('Iterations')
plt.ylabel('Reward')
```

```python
plt.subplot(211)
plt.plot(np.transpose(store_state[str(idx)])[0], 'b')
plt.plot(np.transpose(store_state[str(idx)])[0], 'b.')
plt.ylabel('X-State Error')
plt.title('State Estimation Error')
plt.subplot(212)
plt.plot(np.transpose(store_state[str(idx)])[1], 'b')
plt.plot(np.transpose(store_state[str(idx)])[1], 'b.')
plt.ylabel('Y-State Error')
plt.xlabel('Iterations')
```

```python
plt.subplot(311)
plt.plot(np.transpose(store_meas[str(idx)])[0], 'b')
plt.plot(np.transpose(store_meas[str(idx)])[0], 'b.')
plt.ylabel('TX-1')
plt.title('Multipath affected Observations')
plt.subplot(312)
plt.plot(np.transpose(store_meas[str(idx)])[1], 'b')
plt.plot(np.transpose(store_meas[str(idx)])[1], 'b.')
plt.ylabel('TX-2')
plt.subplot(313)
plt.plot(np.transpose(store_meas[str(idx)])[2], 'b')
plt.plot(np.transpose(store_meas[str(idx)])[2], 'b.')
plt.ylabel('TX-3')
plt.xlabel('Iterations')
```
