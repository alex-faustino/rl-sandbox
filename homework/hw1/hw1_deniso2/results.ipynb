# HW1 part 1
##### import gym and local environments

```python
import gym
import matplotlib.pyplot as plt
import numpy as np
import myenv

dir(myenv) # check which env available
```

##### create env instance, reset it and render

```python
env = myenv.GridWorldEnv()
env.reset()
env.render()
```

##### run a loop with random actions

```python
history = []
for i in range(100):
    a = env.action_space.sample()
    o,r,_,_ = env.step(a)
    env.render()
    print('loc:',o,',rew:',r)
    history.append([i,o[0],o[1],a,r])
```

```python
# plot the history
h = np.asarray(history)
plt.subplot(2, 1, 1)
plt.plot(h[:,0], h[:,4], 'o-')
plt.ylabel('reward')

plt.subplot(2, 1, 2)
plt.plot(h[:,0], h[:,3], '.-')
plt.xlabel('step')
plt.ylabel('action')

plt.show()
```

##### the portals may also be specified manually if needed

```python
env.reset(A=[0,0],Ap=[4,4],B=[4,0],Bp=[0,4])
env.render()
```

# HW1 part 2
##### import gym and local environments

```python
import gym, myenv, time
import numpy as np
import matplotlib.pyplot as plt

dir(myenv)
env = myenv.AcrobotEnv()
```

```python
print(env.reset())
env.render()
history = []

for i in range(20):
    a = env.action_space.sample()
    o,r,_,_ = env.step(a)
    env.render()
    print('obs:',o,',rew:',r)
    history.append([i,o[0],o[1],a,r])
```

```python
# plot the history
h = np.asarray(history)
plt.subplot(3, 1, 1)
plt.plot(h[:,0], h[:,1], 'o-')
plt.ylabel('theta1')

plt.subplot(3, 1, 2)
plt.plot(h[:,0], h[:,2], '.-')
plt.xlabel('step')
plt.ylabel('theta2')

plt.subplot(3, 1, 3)
plt.plot(h[:,0], h[:,4], '.-')
plt.xlabel('step')
plt.ylabel('reward')

plt.show()
```
