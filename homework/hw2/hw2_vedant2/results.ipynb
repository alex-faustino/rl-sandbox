# AE Reinforcement learning - Homework 2

All agents have a linear or exponential annealing to them. The following is the parameters to both SARSA and QLearn agent (with defult initializations):

- def SARSA(env,initial_epsilon = 1, final_epsilon = 0.01,total_episodes = 1000, annealing_period = None,max_steps = 25,lr_rate = 0.9,gamma = 0.9, decay_rate = None):

- def QLearn(env,initial_epsilon = 1, final_epsilon = 0.01,total_episodes = 1000, annealing_period = None,max_steps = 25,lr_rate = 0.9,gamma = 0.9, decay_rate = None):

if decay_rate is not none, exponential annealing model is used with the following formula:

epsilon = initial_epsilon + (final_epsilon - initial_epsilon) * np.exp(-decay_rate * episode) 

## SARSA

SARSA agent for Gridworld

### GridWorld (easy)

```python
# -*- coding: utf-8 -*-
"""
Created on Wed Sep  5 20:30:50 2018

@author: Vedant
"""
import Agents.discreteAgent as DA
import gym
import gridWorld
import numpy as np


env = gym.make('GridWorld-v0')
#env.render('human')
Q, rewards = DA.SARSA(env,initial_epsilon = 1, final_epsilon = 0.01,total_episodes = 1000)
V = np.amax(Q, axis=1)
V = V.reshape(5,5)
pi = np.argmax(Q, axis=1)
pi = pi.reshape(5,5)
print("learnt Q")
print(V)
print("learnt policy")
print(pi)
```

```python
%matplotlib inline

from matplotlib.pyplot import plot

plot(rewards);
```
