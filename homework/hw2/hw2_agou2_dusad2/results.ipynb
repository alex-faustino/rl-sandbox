```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sarsa import Sarsa
from q_learning import Q_Learning
from gridworld import GridWorld
from tqdm import tqdm
```

```python
def plot_state_values(state_dict):
    grid = np.zeros(25).reshape((5,5))
    for i in range(1, 6):
        for j in range(1, 6):
            key = "[{0} {1}]".format(i, j)
            if key in state_dict.keys():
                grid[i-1, j-1] = np.mean(list(state_dict[key].values()))
    
    ax = sns.heatmap(grid)
    plt.show()
```

# SARSA

SARSA GridWorld Easy

```python
sarsa_easy = Sarsa(GridWorld(is_hard=False), step_size=0.1, epsilon=0.1, discount=0.9, episode_length=100)
rewards = []
for i in tqdm(range(100)):
    reward = sarsa_easy.run_episode()
    rewards.append(reward)
plt.figure()
plt.xlabel("Episode")
plt.ylabel("Total reward")
plt.plot(rewards)
plt.show()
```

```python
states = sarsa_easy.test()
print([str(state) for state in states])
print(sarsa_easy.env.show_trajectory(states))
```

```python
plot_state_values(sarsa_easy.Q._states)
```

SARSA GridWorld Hard

```python
sarsa_hard = Sarsa(GridWorld(is_hard=True), step_size=0.1, epsilon=0.1, discount=0.9, episode_length=100)
rewards = []
for i in tqdm(range(100)):
    reward = sarsa_hard.run_episode()
    rewards.append(reward)
plt.figure()
plt.xlabel("Episode")
plt.ylabel("Total reward")
plt.plot(rewards)
plt.show()
```

```python
plot_state_values(sarsa_hard.Q._states)
```
