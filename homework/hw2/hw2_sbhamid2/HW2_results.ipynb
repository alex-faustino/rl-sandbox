__Your goal is to implement two discrete-state/discrete-action reinforcement learning algorithms: SARSA with epsilon-greedy policy (S&B 6.4) and Q-Learning (S&B 6.5).__

__Implement both algorithms so that they can be used on arbitrary discrete-state/discrete-action gym environments. For example, query the environment for the number of states and actions. You will need another parameter which is episode_length, giving the number of timesteps to run the environment before restarting. Apply both algorithms to your grid_world example from HW1. Choose an exploration probability epsilon, learning rate alpha, and episode length so that the training converges. You might need to choose different parameters for each algorithm. Test your algorithms on both the "easy" version of grid_world (actions deterministically move the agent) and the "hard" version (some probability of moving in a direction different to the desired action).__

```python
import gym
import matplotlib.pyplot as plt
import numpy as np
from gridworld.gridworld import GridWorldEnvNew
import qlearning as ql
import SARSAlearning as sa

myenv = gym.make('MyGridworld-v2')
set_params = {}
myenv.seed(543)
```

<font size=5>__Easy version__</font>

```python
##### First set of experiments are conducted for easy version
myenv.env.version = 'easy'
```

```python
## Plotting the learning curves of Qlearning
ql_agent_easy = ql.Qlearning(myenv)
set_params.update({'no_eps': 300, 'eps_len': 150, 'alpha': 0.05, 'gamma': 0.9, 'epsilon': 0.9, 'decay': 0.98})
ql_qfunc_easy, ql_reward_easy = ql_agent_easy.train(set_params)
```

```python
## Plotting the learning curves of SARSA
sarsa_agent_easy = sa.SARSAlearning(myenv)
set_params.update({'no_eps': 300, 'eps_len': 150, 'alpha': 0.05, 'gamma': 0.9, 'epsilon': 0.9, 'decay': 0.98})
sarsa_qfunc_easy, sarsa_reward_easy = sarsa_agent_easy.train(set_params)
```

```python
plt.plot(np.arange(ql_agent_easy.no_eps), ql_reward_easy, 'b', label='Qlearning')
plt.plot(np.arange(sarsa_agent_easy.no_eps), sarsa_reward_easy, 'r', label='SARSA')
plt.legend()
plt.ylabel('Average reward per episode')
plt.xlabel('Episode number')
plt.title('Easy version')
plt.show()
```

```python
### Q learning: example trajectory of robot using trained policy via Q learning
eps_len = 200
record_pos, record_action = ql_agent_easy.test(eps_len)
plt.plot(np.arange(eps_len), record_pos, 'b')
plt.plot(np.arange(eps_len), record_pos, 'b.')
plt.show()
```

```python
### SARSA learning: example trajectory of robot using trained policy via SARSA learning
eps_len = 200
record_pos, record_action = sarsa_agent_easy.test(eps_len)
plt.plot(np.arange(eps_len), record_pos, 'b')
plt.plot(np.arange(eps_len), record_pos, 'b.')
plt.show()
```
