```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
%matplotlib inline

file_name = 'PolicyTransfer/pendNoTransfer_Results'
with open(file_name,'rb') as file:
    avgreward_source = pickle.load(file)

file_name = 'PPO_Pendulum/pendStandalone_Results1'

with open(file_name,'rb') as file:
    avgreward_ppo = pickle.load(file)
    
file_name = 'PolicyTransfer/pendTransfer_Results_1'
with open(file_name,'rb') as file:
    avgreward_transfer1 = pickle.load(file)
    
file_name = 'PolicyTransfer/pendTransfer_Results_2'
with open(file_name,'rb') as file:
    avgreward_transfer2 = pickle.load(file)
        
file_name = 'PolicyTransfer/pendTransfer_Results_3'
with open(file_name,'rb') as file:
    avgreward_transfer3 = pickle.load(file)

fig_1 = plt.figure()    
p1, = plt.plot(avgreward_ppo, label='ppo')
t1, = plt.plot(avgreward_transfer2, label='PT-adap1')
t2, = plt.plot(avgreward_transfer3, label='PT-adap2')
s1, = plt.plot(avgreward_source, label='No-adap')
plt.grid(True)
plt.xlabel('Epoch')
plt.ylabel('Avg Reward')
plt.legend(handles=[p1, t1, t2, s1])
plt.show

fig_1.savefig('Pendulum_Transfer.png')
```
